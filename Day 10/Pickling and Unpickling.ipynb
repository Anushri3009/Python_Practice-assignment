# Pickling and Unpickling

```python
import seaborn as sns
```

```python
dir(sns)
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
import pickle
```

```python
filename='test.pkl'
```

```python
##serialize process
pickle.dump(df,open(filename,'wb'))
```

```python
pickle.load(open(filename,'rb'))
```
